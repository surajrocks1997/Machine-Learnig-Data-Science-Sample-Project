```python
x= [1,2,3,4]
y= [11,22,33,44]
plt.plot(x,y);
```

```python
# plt.plot can also be shown using following command

#1st method
fig = plt.figure()    #creates a figures
ax = fig.add_subplot()    #adds some axes
plt.show()
```

```python
#2nd method
fig = plt.figure()    #creates a figures
ax = fig.add_axes([1,1,1,1])    #adds axes to figure
ax.plot(x,y)
plt.show()
```

```python
#3rd method (RECOMMENDED)
fig, ax = plt.subplots()
ax.plot(x,[50,100,250,300]);
```
