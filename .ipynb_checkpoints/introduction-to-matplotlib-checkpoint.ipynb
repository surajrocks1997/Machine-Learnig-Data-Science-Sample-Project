## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(5,5))    #(width, height)

# 3. plot the data
ax.plot(x,y)

# 4. customize plot
ax.set(title= 'Simple Plot', xlabel= 'x-axis', ylabel= 'y-axis')

# 5. Save and show (save the whole figure)
fig.savefig('images/sample-plot.png')
```

```python
# making plot from dictionary
nut_butter_prices = {'Almond butter': 10,
                    'Peanut butter': 8,
                    'Cashew butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title= "Dan's Nut Butter Prices", 
      ylabel = 'Price ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots


```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10,5))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

#plot to each difference axis (different index)
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib oo method)

* When plotting something quickly, it is okay to use pyplot method.
* When plotting something advanced, use Oo method.


```python
heart_diseas.head()
```

```python
over_50 = heart_diseas[heart_diseas['age'] > 50]
len(over_50)
```

```python
# pyplot method
over_50.plot(kind='scatter', 
             x = 'age', 
             y='chol', 
             c='target');
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45,100])
```

```python
## OO method from scratch
fix, ax = plt.subplots(figsize=(10,6))

# plot data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);

# customize
ax.set(title='Heart Diseas and Cholesterol Levels', 
      xlabel='Age',
      ylabel='Cholesterol');

# add a legent
ax.legend(*scatter.legend_elements(), title='target');

# add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize=(10,10),
                              sharex=True)

# add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

#customize ax0
ax0.set(title='Heart Dieseas and Cholesterol Level',
       ylabel='Cholesterol');

#add a legent to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add mean line
ax0.axhline(over_50['chol'].mean(),
          linestyle = '--');

# add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])

#customize ax1
ax1.set(title='Heart Dieseas and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate');

#add a legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

#add a mean line
ax1.axhline(over_50['thalach'].mean(),
           linestyle='--');


# add a title to a figure
fig.suptitle('Heart Dieseas Analysis', fontsize = 16, fontweight='bold');
```
